```python
def replace_outliers(df):
    for column in df.select_dtypes(include=['number']).columns:  # Fix syntax
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)  # Fix Q3 (previously written as Q2)
        IQR = Q3 - Q1  # Corrected IQR calculation
        lb = Q1 - 1.5 * IQR  # Lower bound
        ub = Q3 + 1.5 * IQR  # Upper bound

        df[column] = df[column].apply(lambda x: Q1 if x < lb else (Q3 if x > ub else x))  # Fix column assignment

    return df
```

```python
df=replace_outliers(df)
```
